```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
my_original_img = cv2.imread("./aslan.jpg",0)
```

```python
# _ simgesi python dilinde gereksiz, işe yaramayan bir değerin tutulması
# için kullanılır. Bu örnekte, cv.thresholding() fonksiyonu 2 değer
# döndürür. bu değerlerden ilki thresh yani eşik değeridir.
# biz bu değeri kullanmayacağımız için _ simgesine aktarırız.
# Thresholding (eşikleme) işlemi, görüntü işlemede 0-255 değerlerinden oluşan
# bir görüntünün piksellerinin değerini yeniden belirler.
# Burada, thresh yani eşik değeri bir limit belirtir.
# ilgili piksel eğer bu eşik değerinin altındaysa, üstündeyse ya da eşitse
# thresh tekniğine göre farklı bir değer alır.
# Dolayısıyla gri seviye bir görüntü farklı piksel değerleriyle yeniden
# temsil edilir thresholding işlemi sayesinde.

# Binary thresholding (ikili eşikleme)
# görüntüdeki ilgili pikel thresh değerinin altındaysa sıfır değerini alır
# yani siyah renge dönüştürülür, thresh değerinin üstündeyse 255 değerini
# alır, yani beyaz renge dönüştürülür ( ya 0 ya da 255, yani binary-ikili)

_, my_thresh_binary_img = cv2.threshold(my_original_img, thresh=128,
            maxval=255, type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")
```

```python
_, my_thresh_binary_inv_img = cv2.threshold(my_original_img, thresh=128,
            maxval=255, type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDING Image")
```

```python
# Tozero Thresholding
# Piksel değeri eşik değerinden büyükse, pikselin değeri olduğu gibi bırakılır. 
# Ancak, eşik değerinden küçük veya eşitse, o pikselin değeri 0 yapılır.

_, my_thresh_tozero_img = cv2.threshold(my_original_img, thresh=128,
            maxval=255, type=cv2.THRESH_TOZERO)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDING Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_tozero_img, cmap="gray")
plt.axis('off')
plt.title("TOZERO THRESHOLDING Image")
```
