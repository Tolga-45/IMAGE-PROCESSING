```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# aslan.jpg görüntüsünü renkli bir şekilde okuyun
# aşağıdaki kod ile bir görüntüyü BGR kanal sıralamasıyla renkli bir
# şekilde 3 kanallı olarak okur

my_img = cv2.imread("./aslan.jpg")
```

```python
plt.imshow(my_img)
```

```python
my_img_2 = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(my_img_2)
```

```python
# Flipping (Çevirme) İşlemleri
# Görüntüyü x, y ya da hem x hem de y eksenlerinde 180 derece döndürür

# cv2.flip() fonksiyonuyla çevirme işlemleri yapılır
# kullanımı=> cv2.flip(görüntü, numara)

# cv2.flip(görüntü, 0) => dikey eksende 180 derece döndürür
# cv2.flip(görüntü, 1) => yatay eksende 180 derece döndürür
# cv2.flip(görüntü, -1) => hem dikey hem de yatay eksende 180 derece döndürür

my_img_vertical = cv2.flip(my_img_2, 0)
my_img_horizontal = cv2.flip(my_img_2, 1)
my_img_both = cv2.flip(my_img_2, -1)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2)
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_img_vertical)
plt.axis('off')
plt.title("VERTICAL Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_img_horizontal)
plt.axis('off')
plt.title("HORIZONTAL Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_img_both)
plt.axis('off')
plt.title("VERTICAL and HORIZONTAL Image")
```

```python
# Resize: bir görüntünün yükseklik, genişlik ya da her ikisinin yeniden
# boyutlandırılmasıdır.

# Kullanımı: cv2.resize(görüntü_değişkeni,(yeni genişlik, yeni yükseklik),
# interplasyon_yöntemi)
print(f"Görüntünün Boyut Bilgileri: {my_img_2.shape}")
print("-"*50)
print(f"Görüntünün yükseliği = {my_img_2.shape[0]}")
print(f"Görüntünün genişliği = {my_img_2.shape[1]}")
```

```python
my_resized_img = cv2.resize(my_img_2, (600,450), cv2.INTER_LANCZOS4)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2)
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_resized_img)
plt.axis('off')
plt.title("RESIZED Image")
```
