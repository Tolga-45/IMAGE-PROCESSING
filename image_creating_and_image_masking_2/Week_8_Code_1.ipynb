```python
# Kod ile kırmızı rengini oluşturma

width, height , channel = 300, 300, 3

img_red = np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için RGB kanal sıralaması değerleri = [255,0,0]
img_red[:,:,0] = 255

plt.imshow(img_red)
plt.axis('off')
```

```python
# Kod ile yeşil rengini oluşturma

width, height , channel = 300, 300, 3

img_green = np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için RGB kanal sıralaması değerleri = [255,0,0]
img_green[:,:,1] = 255
```

```python
# Kod ile mavi rengini oluşturma

width, height , channel = 300, 300, 3

img_blue = np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için RGB kanal sıralaması değerleri = [255,0,0]
img_blue[:,:,2] = 255
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_red)
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_green)
plt.axis('off')
plt.title("GREEN Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_blue)
plt.axis('off')
plt.title("BLUE Image")
```

```python
# Siyah renk çizdirme

img_black = np.zeros((width,height,3),dtype=np.uint8)
img_white = np.ones((width,height,3),dtype=np.uint8)*255

img_gray_1 = np.zeros((width,height,3),dtype=np.uint8)
img_gray_1[:,:,:] = 128

img_gray_2 = np.random.randint(128,129,(width,height,3),dtype=np.uint8)
```

```python
# Sarı renk:
# RGB sıralaması = [255,255,0]
# Red: 255
# Green: 255
# Blue: 0
```

```python
width, height, channel = 300, 420, 3

img_flag_germany = np.zeros((width,height,channel),dtype=np.uint8)

# 0:100 demek ilk 100 satır demektir
img_flag_germany[:100,:] = [0,0,0] # RGB = [0,0,0]

# 100:200 demek 100. satırdan 200. satıra kadar, anlamına gelir
img_flag_germany[100:200,:] = [255,0,0] # RGB = [255,0,0]

# 200:300 demek 200. satırdan 300. satıra kadar, anlamına gelir
img_flag_germany[200:300,:] = [255,255,0] # RGB = [255,255,0]

plt.imshow(img_flag_germany)
plt.axis('off')
```

```python
# Belçika Bayrağı
width, height, channel = 300, 420, 3

img_flag_belgium = np.zeros((width,height,channel),dtype=np.uint8)

# 0:140 demek ilk 140 sütun demektir
img_flag_belgium[:,:140] = [0,0,0] # RGB = [0,0,0]

# 140:280 demek ilk 140. sütundan 280. sütuna kadar demektir
img_flag_belgium[:,140:280] = [255,255,0] # RGB = [255,0,0]

# 280:420 demek ilk 280. sütundan 420. sütuna kadar demektir
img_flag_belgium[:,280:420] = [255,0,0] # RGB = [255,255,0]

plt.imshow(img_flag_belgium)
plt.axis('off')
```

```python
width, height, channel = 300, 420, 3

img_flag_russia = np.zeros((width,height,channel),dtype=np.uint8)

# 0:100 demek ilk 100 satır demektir
img_flag_russia[:100,:] = [220,220,220] # RGB = [0,0,0]

# 100:200 demek 100. satırdan 200. satıra kadar, anlamına gelir
img_flag_russia[100:200,:] = [0,0,255] # RGB = [255,0,0]

# 200:300 demek 200. satırdan 300. satıra kadar, anlamına gelir
img_flag_russia[200:300,:] = [255,0,0] # RGB = [255,255,0]

plt.imshow(img_flag_russia)
plt.axis('off')
```

```python
# Var olan bir rengi başka bir renge boyamak işlemine MASKELEME (MASKING) denir
# Maskeleme için önce değiştirmek istediğiniz rengin, kanallara göre piksel
# değerini elde edersiniz

# Örneğin, bir görüntüde sadece ama sadece kırmızı rengi seçmek istiyorsanız
# Maskeleme işleminde kırmızı renge karşılık gelen piksel değerlerini
# 3 Kanal için de seçersiniz ve bu sizin maskeniz olur.
# İkinci aşamada, değiştirmek istediğiniz rengin piksel değerlerini hazırlarsınız
# original_img[maskelenmiş_pikseller] = [yeni pikseller]

# şeklinde bir atama işlemi ile artık iki rengi değiştirirsiniz.
# RGB KANAL SIRALAMASINA GÖRE KIRMIZI RENK = [255,0,0]

# mask_red değişkeni, bir görüntüdeki bütün kırmızı renkleri seçen maskeleyen koddur
mask_red = (img_flag_russia[:, :, 0] == 255) & (img_flag_russia[:, :, 1] == 0) 
& (img_flag_russia[:, :, 2] == 0)

# aşağıdaki kod, bir görüntüde artık sadece kırmızı renkleri göstermek
# için verilmiştir.

img_flag_russia[mask_red] = [255, 255, 0]

plt.imshow(img_flag_russia)
plt.axis('off')
```
